Class for handling sparse matrices

```python
class SparseMatrices:
    def __init__(
        self,
        alt_mtx, 
        ref_mtx,
        index:list,
        columns:list
    ):  
        # checks        
        if alt_mtx.shape != (len(index), len(columns)):
            raise ValueError("Problem with index and columns")
        
        self.alt_mtx = alt_mtx.tocsr()
        self.ref_mtx = ref_mtx.tocsr()
        self.index = {name: idx for idx, name in enumerate(index)}
        self.columns = {name: idx for idx, name in enumerate(columns)}
        self.cov_mtx = alt_mtx + ref_mtx
        self.alt_frac = alt_mtx.multiply(self.cov_mtx.power(-1)).tocsr()
           
    def mask_alt_frac(self, cov_thresh:int):
        """
        mask alt_frac where coverage < cov_thresh
        """
        masks = self.cov_mtx < cov_thresh
        self.alt_frac = self.alt_frac.multiply(masks.astype(float))
        
    def subset_matrix(attr_name):
        def decorator(func):
            def wrapper(self, index=None, columns=None):
                matrix = getattr(self, attr_name)

                # Row indexing
                if index is None:
                    row_idx = slice(None)
                elif isinstance(index[0], bool):
                    row_idx = index
                else:
                    row_idx = [self.index[i] for i in index]

                # Column indexing
                if columns is None:
                    col_idx = slice(None)
                elif isinstance(columns[0], bool):
                    col_idx = columns
                else:
                    col_idx = [self.columns[c] for c in columns]

                return matrix[row_idx, :][:, col_idx]

            return wrapper
        return decorator

    @subset_matrix("alt_frac")
    def subset_alt_frac(self): pass

    @subset_matrix("cov_mtx")
    def subset_cov_mtx(self): pass

    @subset_matrix("alt_mtx")
    def subset_alt_mtx(self): pass

    @subset_matrix("ref_mtx")
    def subset_ref_mtx(self): pass
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

def plot_scatter(
    data, x, y, hue=None, palette=None,
    width=1.22*4, height=0.94*4,
    dot_size=1, filename="scatter.svg",
    alpha=1, show_legend=False
):
    fig, ax = plt.subplots(figsize=(width, height))

    # Assign colors
    if hue:
        colors = data[hue].map(palette).fillna("#D4D4D4").values
    else:
        colors = "#D4D4D4"

    # Plot points
    ax.scatter(
        data[x],
        data[y],
        c=colors,
        s=dot_size,
        alpha=alpha,
        edgecolor="none"
    )

    # Hide axes
    for spine in ax.spines.values():
        spine.set_visible(False)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_facecolor("white")
    fig.patch.set_facecolor("white")

    # Add legend if hue is used
    if show_legend:
        legend_elements = [
            Patch(facecolor=color, label=label, edgecolor="none")
            for label, color in palette.items()
            if label in data[hue].unique()
        ]
        ax.legend(
            handles=legend_elements, 
            title=hue, 
            loc="center right", 
            bbox_to_anchor=(1.2, 0.5),
            fontsize=8, 
            title_fontsize=9
        )

    # Save
    plt.savefig(filename, bbox_inches="tight", pad_inches=0)
```

#### Exploratory Analysis

Get homozygous loci

```python
wildcards = SimpleNamespace()
wildcards.species = "ptr"
hom_ref = VariantFile(path_hom_ref(wildcards))
hom_loci = [f"{i.chrom}:{i.pos}" for i in hom_ref.fetch()]
```

```python
# read matrices
matrix_dict = {}
for batch, sample in batch_name.items():
    if batch in SKIPPED_BATCH:
        continue
    # wildcards
    wildcards = SimpleNamespace()
    wildcards.species = "ptr"
    wildcards.sample = sample
    
    # 
    sparse_matrices = SparseMatrices(
        alt_mtx=mmread(path_alt_mtx(wildcards)),
        ref_mtx=mmread(path_ref_mtx(wildcards)),
        index=[i.strip() for i in open(path_snv_loci(wildcards), "r")],
        columns=[f"{i.strip()}_{batch}" for i in open(path_barcodes(wildcards), "r")]
    )
    
    matrix_dict[sample] = sparse_matrices
```

PCA and UMAP of SNV per sample

```python
ipca = IncrementalPCA(n_components=N_PCS)
for sample, matrix in matrix_dict.items():
    # setup umap transformer
    umap = UMAP(**UMAP_PARAMS)
    
    # get alt_frac at hom_ref
    hom_index = list(set(hom_loci) & set(matrix.index.keys()))
    hom_index.sort(key=lambda x: matrix.index[x])
    hom_alt_frac = matrix.subset_alt_frac(index=hom_index)
    fig = sns.displot(
        matrix.subset_cov_mtx(index=hom_index).data,
        kind="hist",
        binwidth=1,
        binrange=(0,50)
    )
    fig.savefig(path_output / f"{sample}_alt_frac_hist.svg")
    
    # mask hom-alt-frac where cov-mtx < 5
    hom_cov_mtx_gt5 = matrix.subset_cov_mtx(index=hom_index) >= COV_THRESHOLD
    hom_alt_frac_masked = hom_alt_frac.multiply(hom_cov_mtx_gt5)
    
    # subset only loci (rows) with non-zero alt_frac
    nonzero_loci_idx = list(hom_alt_frac_masked.nonzero()[0])
    nonzero_loci_idx = set(nonzero_loci_idx)
    nonzero_loci_idx = list(nonzero_loci_idx)
    nonzero_loci_idx.sort()
    tmp = {v:k for k, v in matrix.index.items()}
    nonzero_loci = [tmp[idx] for idx in nonzero_loci_idx]
    
    # subset only cells (columns) with non-zero alt_frac
    nonzero_cell_idx = list(hom_alt_frac_masked.nonzero()[1])
    nonzero_cell_idx = set(nonzero_cell_idx)
    nonzero_cell_idx = list(nonzero_cell_idx)
    nonzero_cell_idx.sort()
    tmp = {v:k for k, v in matrix.columns.items()}
    nonzero_cell = [tmp[idx] for idx in nonzero_cell_idx]
    
    # subset
    hom_alt_frac_masked_nonzero = hom_alt_frac_masked[nonzero_loci_idx, :][:, nonzero_cell_idx]
    
    # pca
    pca_projection = ipca.fit_transform(hom_alt_frac_masked_nonzero.T)
    
    # umap
    umap_projection = umap.fit_transform(pca_projection)
    
    # plot umap
    data = pd.DataFrame(umap_projection)
    data.index = nonzero_cell
    batch = name_batch[sample]
    data["cluster"] = cell_identity.loc[nonzero_cell, "Cluster"].values
    plot_scatter(data, 0, 1, "cluster", color_palette, filename=path_output / f"umap_{sample}.svg")
    
    # dynamicviz umap
    if os.path.isfile(path_output / f"dynamicviz_{sample}.csv"):
        out = pd.read_csv(path_output / f"dynamicviz_{sample}.csv")
    else:
        X = pca_projection
        Y = data.copy()
        Y["barcode"] = Y.index
        out = boot.generate(
            X=X, method="umap", Y=Y, B=B, no_bootstrap=True, num_jobs=20,
            save=path_output / f"dynamicviz_{sample}.csv",
            use_n_pcs=False,
            subsample=False,
            return_times=False,
            **DYNAMICVIZ_UMAP_PARAMS
        )
    median = out.groupby("barcode").median()
    plot_scatter(median, "x1", "x2", "cluster", color_palette, filename=path_output / f"median-umap_{sample}.svg")
    plot_scatter(out, "x1", "x2", "cluster", color_palette, filename=path_output / f"stacked-umap_{sample}.svg", alpha=0.1)
```

### All sample PCA and UMAP

```python
# setup sparse matrix to record alt_frac from all samples
# row.index = hom_loci
# columns = cells
row_idx = {name: idx for idx, name in enumerate(hom_loci)}
cells = []
for sample, matrix in matrix_dict.items():
    sample_cells = list(matrix.columns.keys())
    sample_cells.sort(key=lambda x: matrix.columns[x])
    cells += sample_cells
col_idx = {name: idx for idx, name in enumerate(cells)}
alt_frac_all_sample = csr_matrix((len(row_idx), len(col_idx)))

stored_elements = 0
for sample, matrix in matrix_dict.items():
    # get alt_frac at homozygous loci
    hom_index = list(set(hom_loci) & set(matrix.index.keys()))
    hom_index.sort(key=lambda x: matrix.index[x])
    hom_alt_frac = matrix.subset_alt_frac(index=hom_index)
    
    # mask hom-alt-frac where cov-mtx < COV_THRESHOLD
    hom_cov_mtx_passed = matrix.subset_cov_mtx(index=hom_index) >= COV_THRESHOLD
    hom_alt_frac_masked = hom_alt_frac.multiply(hom_cov_mtx_passed)
    
    # get data and index
    data = hom_alt_frac_masked.data
    sample_row_idx, sample_col_idx = hom_alt_frac_masked.nonzero()
    
    # covert sample row, col index to names
    sample_row_name = [hom_index[i] for i in sample_row_idx]
    tmp = {idx: name for name, idx in matrix.columns.items()}
    sample_col_name = [tmp[i] for i in sample_col_idx]
    
    # covert sample row, col names to all row, col index
    all_row_idx = [row_idx[name] for name in sample_row_name]
    all_col_idx = [col_idx[name] for name in sample_col_name]

    sample_new_sparse = csr_matrix(
        (data, (np.array(all_row_idx), np.array(all_col_idx))),
        shape=(len(row_idx), len(col_idx))
    )
    alt_frac_all_sample += sample_new_sparse
    stored_elements += len(data)
    if not len(alt_frac_all_sample.data) == stored_elements:
        break
```

Differential RNA editing using REDITs 

```python
not_all_zero_alt_frac = not_all_zero_alt_frac.tocsc()
```

```python
cond1 = [3]
cond2 = [1, 2, 4, 5, 6, 7, 8]
cond1_label = "RO"
cond2_label = "Others"
cond1_cells = cell_identity.loc[cell_identity["Cluster"].isin(cond1), "Barcode"].values.tolist()
cond2_cells = cell_identity.loc[cell_identity["Cluster"].isin(cond2), "Barcode"].values.tolist()
cond1_cells_set = set(cond1_cells)
cond2_cells_set = set(cond2_cells)
    
# 
nonzero_loci_dict = {locus: idx for idx, locus in enumerate(nonzero_loci)}
nonzero_cells_dict = {cell: idx for idx, cell in enumerate(nonzero_cells)}
nonzero_loci_set = set(nonzero_loci)
nonzero_cells_set = set(nonzero_cells)

# for each locus passing threshold test differential editing
cond1_passed = not_all_zero_alt_frac[:, [nonzero_cells_dict[i] for i in list(cond1_cells_set & nonzero_cells_set)]].getnnz(axis=1) / len(cond1_cells_set) > PERCENT_DETECTED
cond2_passed = not_all_zero_alt_frac[:, [nonzero_cells_dict[i] for i in list(cond2_cells_set & nonzero_cells_set)]].getnnz(axis=1) / len(cond2_cells_set) > PERCENT_DETECTED
both_passed = cond1_passed & cond2_passed
both_passed_loci = np.array(nonzero_loci)[both_passed]

for locus in both_passed_loci:
    # collect alt and ref count
    all_cond_cells = cond1_cells + cond2_cells
    read_count = {
        "barcode": [],
        "ref_counts": [],
        "alt_counts": []
    }
    for sample, matrix in matrix_dict.items():
        batch = name_batch[sample]
        all_cond_cells_batch = [i for i in all_cond_cells if i.endswith(f"_{batch}")]
        if (matrix.index.get(locus, False)) & len(all_cond_cells_batch) > 0:
            sample_ref_mtx = matrix.subset_ref_mtx(index=[locus], columns=all_cond_cells_batch)
            sample_alt_mtx = matrix.subset_alt_mtx(index=[locus], columns=all_cond_cells_batch)
            read_count["ref_counts"] += list(sample_ref_mtx.toarray()[0])
            read_count["alt_counts"] += list(sample_alt_mtx.toarray()[0])
            read_count["barcode"] += all_cond_cells_batch
        
    # write to file
    def get_group(x, set1, set2, label1, label2):
        if x in set1:
            return label1
        if x in set2:
            return label2
        print(x)
        return "none"
    read_count_df = pd.DataFrame(read_count)
    read_count_df["coverage"] = read_count_df["ref_counts"] + read_count_df["alt_counts"]
    read_count_df["condition"] = read_count_df["barcode"].apply(lambda x: get_group(x, cond1_cells_set, cond2_cells_set, cond1_label, cond2_label))
    read_count_df_gt_thresh = read_count_df[read_count_df["coverage"] >= COV_THRESHOLD]
    read_count_df_gt_thresh.drop("coverage", axis=1, inplace=True)
    read_count_df_gt_thresh.to_csv(path_output / f"read_counts/{locus}_{cond1_label}_{cond2_label}.csv", index=False)
    
    # plot histogram of editing level
    read_count_df_gt_thresh["editing_level"] = read_count_df_gt_thresh["alt_counts"] / (read_count_df_gt_thresh["ref_counts"] + read_count_df_gt_thresh["alt_counts"])
    read_count_df_gt_thresh["celltype"] = cell_identity.loc[read_count_df_gt_thresh["barcode"], "Cluster"].values
```

Plot editing level on Seurat CCA integrated median UMAP

```python
median_umap = pd.read_csv(path_median_umap)
median_umap.head()
```

```python
import matplotlib as mpl

# Colormap and fixed normalization
cmap = mpl.cm.get_cmap("viridis")
norm = mpl.colors.Normalize(vmin=0, vmax=1)

# ScalarMappable for colorbar
sm = mpl.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])

for row_tuple in p_adj.itertuples():
    if row_tuple.p_adj >= 0.05:
        continue
    # Load and process data
    data = pd.read_csv(path_output / f"read_counts/{row_tuple.locus}")
    data["editing_level"] = data["alt_counts"] / (data["ref_counts"] + data["alt_counts"])

    # Merge with UMAP
    umap_projection = median_umap.copy()
    umap_projection = umap_projection.merge(data, "left", left_on="Barcode", right_on="barcode")

    # Split the data
    df_nan = umap_projection[umap_projection["editing_level"].isna()].copy()
    df_valid = umap_projection[umap_projection["editing_level"].notna()].copy()

    # First: create grid without hue for plotting NaNs
    grid = sns.FacetGrid(
        umap_projection,  # full dataset so FacetGrid layout matches
        col="Batch",
        col_wrap=5,
        height=2.2
    )

    # Plot NaN values in gray (they’ll be in the background)
    grid.map_dataframe(
        sns.scatterplot,
        data=df_nan,
        x="x1",
        y="x2",
        color="#ebebeb",
        s=0.5,
        edgecolor=None,
        legend=False
    )

    # Overlay colored points (valid editing levels)
    for ax, batch_name in zip(grid.axes.flat, grid.col_names):
        subset_valid = df_valid[df_valid["Batch"] == batch_name]
        if not subset_valid.empty:
            sns.scatterplot(
                data=subset_valid,
                x="x1",
                y="x2",
                hue="editing_level",
                hue_norm=(0, 1),
                palette="viridis",
                s=0.8,
                edgecolor=None,
                ax=ax,
                legend=False
            )
    
    # Add colorbar to the right
    cbar = grid.fig.colorbar(
        sm,
        ax=grid.axes,
        orientation="vertical",
        fraction=0.02,
        pad=0.04,
        aspect=40
    )
    cbar.set_label("Editing level")

    # 
    grid.set_titles(col_template="{col_name}")
    grid.savefig(path_output / f"editing_on_median_umap/{row_tuple.locus.rstrip('.csv')}.png", dpi=600)
    plt.close()
```

```python
annotation_synonym = {}
with open(path_annotation_synonym, "r") as handle:
    for line in handle:
        line = line.strip()
        tid, *others = line.split("\t")
        *gid, _ = tid.split(".")
        gid = ".".join(gid)
        annotation_synonym[gid] = ";".join([annotation_synonym.get(gid, ""), tid, *others]).strip(";")
 
annotation_synonym
```
