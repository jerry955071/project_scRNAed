```python
import pysam
from pathlib import Path

# vars
path_work_dir = Path("/home/b05b01002/HDD/project_scRNAed")
bam = pysam.AlignmentFile(path_work_dir / "outputs/CellRanger/count/ptr_tenx_batch1/outs/possorted_genome_bam.bam", "rb")

xf_counts = {}
loop_n = 0
for read in bam:
    loop_n += 1
    if loop_n % 100000 == 0:
        print(f"Processed {loop_n} reads...")
    tags = dict(read.get_tags())
    xf_value = tags.get("xf", "NA")
    if xf_counts.get(xf_value) is None:
        xf_counts[xf_value] = 1
    else:
        xf_counts[xf_value] += 1
```
